## Aproximación de integrales utilizando el método trapezoidal

Suponga que desea aproximar la integral (definida) de una función $f$, dentro del intervalo $[a, b]$

$$
\int_{a}^{b}f(x)dx
$$

El método trapezoidal nos dice que podemos aproximar tal integral utilizando el siguiente algoritmo:

1. Divida el intervalo $[a, b]$ en $n$ segmentos de tamaño $\Delta$

$$
\Delta = \dfrac{b - a}{n}
$$

2. Iniciando con $x_{0} = a$, obtenga los $n + 1$ puntos $x_i$ de la siguiente manera:
  * $x_0 = a$
  * $x_1 = a + \Delta$
  * $x_2 = a + 2 \Delta$
  * $x_3 = a + 3 \Delta$
  * $\vdots$
  * $x_n = a + n \Delta = b$

3. Aproxime la integral utilizando la fórmula

$$
\int_{a}^{b}f(x)dx \approx \Delta \dfrac{f(x_0) + f(x_1)}{2} + \Delta \dfrac{f(x_1) + f(x_2)}{2} + \ldots + \Delta \dfrac{f(x_{n-1}) + f(x_n)}{2}
$$

````python
def funcion_prueba(x):
    return x**2

def crea_particion(lim_inf, lim_sup, num_seg):
    '''
    Función para crear la partición con las xi's
    
    ENTRADA
    lim_inf: Float que representa el límite inferior del intervalo sobre que se integra
    
    lim_sup: Float que representa el límite superior del intervalo sobre que se integra
    
    num_seg: Entero positivo que representa el número de segmentos en los que se divide
    el intervalo de integración
    
    SALIDA
    lista con los puntos xi
    '''
    
    delta = float ((lim_sup - lim_inf))/ num_seg
    
    x = []
    x.append(lim_inf)
    
    for i in range(1, num_seg + 1):
        x.append(lim_inf + i * delta)
        
    return x

def integracion_trapezoidal(f, lim_inf, lim_sup, num_seg):
    '''
    Función para aproximar la integral de una función 'f' utilizando el trapecios
    
    ENTRADA
    f: Nombre de un función

    lim_inf: Float que representa el límite inferior del intervalo sobre que se integra
    
    lim_sup: Float que representa el límite superior del intervalo sobre que se integra
    
    num_seg: Entero positivo que representa el número de segmentos en los que se divide
    el intervalo de integración
    
    SALIDA
    float
    '''
    
    delta = float ((lim_sup - lim_inf))/ num_seg
    
    #Crea los puntos xi
    x = crea_particion(lim_inf, lim_sup, num_seg)
    
    #para almacenar las sumas
    suma = 0
    
    for i in range(num_seg):
        suma = suma + delta * 0.5 * (f(x[i]) + f(x[i + 1]))
    return suma    
````


````python
n = 4
for reng in range(n + 1):
    if reng != 1:
        s = ''
        for col in range(n):
            if reng == 0:
                prod = (reng + 1) * (col + 1)
            else:
                prod = reng * (col + 1)
            if prod < 10:
                s = s + '0' + str(prod)
            else:
                s = s + str(prod)
            if col == 0:
                s = s + '|'
            else:
                s = s + ' '
    else:
        s= '-'*n*3
        s = s[0:(len(s) - 1)]
    print(s)
````

# Ejercicio 4

Escriba una función `fibonacci` que regrese una lista con los primeros `n` elementos de la secuencia Fibonacci. Recordemos, la secuencia fibonacci es la siguiente:
```
0 1 1 2 3 5 8 13 21
```
En general, el $n$-ésimo término de la secuencia Fibonacci es $F_1 = 0$, $F_2 = 1$ y, para cualquier $n \geq 3$,

$$
    F_n = F_{n - 2} + F_{n - 1}
$$

````python
def fibonacci(n):
    if n==1:
        return [1]
        #return[0]
    elif n==2:
        return [1,1]
        #return [0, 1]
    else:
        F = [0]*n
        F[0] = 1 #F[0] = 0
        F[1] = 1 #F[1] = 1
        
        for i in range(2, n):
            F[i] = F[i - 1] + F[i - 2]
    return F     
````
